```python
import os,glob
import xarray
import wrf
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
ncpath_1 = "/glade/scratch/doubrawa/postProcessing/WRF_LES_100m_100.0_m_AGL_2015-03-21_14:01.nc"
ncfile_1 = xarray.open_dataset(ncpath_1)

ncpath_2 = "/glade/scratch/doubrawa/postProcessing/WRF_LES_25m_100.0_m_AGL_2015-03-21_14:00.nc"
ncfile_2 = xarray.open_dataset(ncpath_2)
```

```python
time_smart = ncfile_1.time[0].data
```

```python
lat2d_1 = ncfile_1.lat
lon2d_1 = ncfile_1.lon

lat2d_2 = ncfile_2.lat
lon2d_2 = ncfile_2.lon
```

```python
values_1 = ncfile_1.w[0,:,:]
values_2 = ncfile_2.w[0,:,:]

vmin = -1
vmax = 1
dv   = 0.2

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
p1 = ax1.contourf(wrf.to_np(lon2d_1), wrf.to_np(lat2d_1), wrf.to_np(values_1), levels=np.arange(vmin,vmax,dv))
ax1.set_title("d03")

ax2 = fig.add_subplot(1,2,2,sharex=ax1,sharey=ax1)
p2 = ax2.contourf(wrf.to_np(lon2d_2), wrf.to_np(lat2d_2), wrf.to_np(values_2), levels=np.arange(vmin,vmax,dv))
ax2.set_title("d04")

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.82, 0.255, 0.02, 0.49])
clb = fig.colorbar(p2, cax=cbar_ax)
```
