```python
import xarray
import os, glob
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
main_path = "/glade/scratch/doubrawa/post_processing"
```

```python
ncfiles = sorted(glob.glob(os.path.join(main_path,"WRF_LES_25m_3D_*0.nc")))
```

```python
for ncfile in ncfiles:
    data = xarray.open_dataset(ncfile)
    try:
        data_little = data.isel(bottom_top=6)
        ws = data_little.u**2 
        print("{0:60s} {1:.1f} m2/s2".format(os.path.basename(ncfile),np.mean(ws.values)))
    except:
        print("no u in {0}".format(os.path.basename(ncfile)))
```

```python
csvfiles = glob.glob(os.path.join(main_path,"*.csv"))
```
