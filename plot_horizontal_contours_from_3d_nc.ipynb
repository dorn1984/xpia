```python
import glob, os
import xarray
from scipy import fftpack
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
fs=14
plt.rc("font",size=fs)
```

```python
from matplotlib.ticker import FormatStrFormatter
```

```python
inpath = "/glade/scratch/doubrawa/post_processing/"
```

```python
filepaths = sorted(glob.glob(os.path.join(inpath,"*_plus_filtered.nc")))
```

```python
# 137

filepaths[78]
```

```python
for filepath in filepaths:
    data = xarray.open_dataset(filepath)
    datetime = pd.to_datetime(data.valid,format="%Y-%m-%d_%H:%M")
    approx_z = data.z.mean(dim="west_east").mean(dim="south_north").data
    print(filepath+"k = 0 ~ {0:.1f} m".format(approx_z[0]))
```

```python
approx_z = data.z.mean(dim="west_east").mean(dim="south_north").data
```

```python
i1 = 400
i2 = 800

vmin = -2
vmax = 2

for k in [5,8,11,14,17]:#np.arange(5,8,2):

    data_hor_slice = data.isel(bottom_top=k)
    data_hor_slice = data_hor_slice.isel(south_north=np.arange(i1,i2,1)).isel(west_east=np.arange(i1,i2,1))
    
    fig = plt.figure(figsize=(6,6))
    ax  = fig.add_subplot(111, aspect='equal')
    p = ax.pcolormesh(data_hor_slice.west_east,data_hor_slice.south_north,data_hor_slice.w, vmin=vmin, vmax=vmax, cmap='RdBu')
    ax.set_title("$z\sim{0:.0f}$ m".format(approx_z[k]))

    desired_ticks = [10,15,20]
    xticks        = [(ii*1000)/25.0 for ii in desired_ticks]
    labels        = ["{0:.0f}".format(dx) for dx in desired_ticks]    
    
    ax.set_title("{0:%Y-%m-%d %H:%M} UTC\n$z\sim{1:.0f}$ m".format(datetime,approx_z[k]),fontsize=fs,y=1.01)
    ax.set_xticks(xticks)
    ax.set_xticklabels(labels,fontsize=fs)
    ax.set_yticks(xticks)
    ax.set_yticklabels(labels,fontsize=fs) 
    ax.set_xlabel("East-West Distance [km]",labelpad=10)
    ax.xaxis.set_tick_params(direction='in')
    ax.yaxis.set_tick_params(direction='in')     
    ax.tick_params(axis='both', which='major', pad=10)
    
    clb = plt.colorbar(p,format=FormatStrFormatter('%4.1f'),shrink=0.8)
    clb.set_label('Vertical Velocity '+r'$w$ [m s$^{{-1}}$]', labelpad=10, fontsize=17) 
    
    fig.savefig("/glade/u/home/doubrawa/figs/w_k_{0:.0f}_t_{1:%Y-%m-%d_%H-%M}.png".format(k,datetime),
               bbox_inches="tight")
```
