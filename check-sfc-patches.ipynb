```python
import xarray
import matplotlib.pyplot as plt
```

```python
plt.rc("font",size=14)
```

```python
import os, glob
```

```python
maindir = "/glade/scratch/doubrawa/post_processing"
```

```python
fpaths = sorted(glob.glob(os.path.join(maindir,"WRF_LES_25m_SFC_*")))
```

```python
fpaths
```

### i/o

```python
for fpath in fpaths:
    xr = xarray.open_dataset(fpath)

    xlat    = xr.lat
    xlon    = xr.lon
    x       = xr.west_east
    y       = xr.south_north
    contour = 1/(xr.RMOL)

    ny, nz = xr.HFX.shape

    box_side_km = 10.0
    box_side_m  = box_side_km * 1000.0 
    delta       = 25.0 
    box_npts    = box_side_m / delta
    npts_fringe = (ny-box_npts)/2
    start       = int(npts_fringe)
    end         =  int(npts_fringe) + int(box_npts)

    xlat_subset    = xlat[start:end,start:end]
    xlon_subset    = xlon[start:end,start:end]
    x_subset       = x[start:end]
    y_subset       = y[start:end]
    contour_subset = contour[start:end,start:end]

    datetime_str = os.path.split(fpath)[-1].split('_SFC_')[-1].split('.nc')[0]

    fig = plt.figure(figsize=(9,9))
    ax  = fig.add_subplot(111)
    p1 = ax.pcolormesh(x_subset,y_subset,contour_subset)#,vmin=-100,vmax=100)
    ax.set_xlabel("west_east points")
    ax.set_ylabel("south_north points")
    ax.set_title(datetime_str)
    plt.colorbar(p1)
    fig.savefig("/glade/u/home/doubrawa/figs/1_over_RMOL_v2_{0}".format(datetime_str),dpi=330,bbox_inches="tight")
    plt.clf()
    plt.close()
```
